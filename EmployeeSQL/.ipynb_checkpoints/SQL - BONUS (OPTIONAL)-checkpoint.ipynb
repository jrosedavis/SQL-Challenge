```python
# Dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sqlalchemy
import psycopg2
from matplotlib.ticker import StrMethodFormatter
```

```python
from sqlalchemy import create_engine
#Connection is how you connect to your SQL database
engine = create_engine('postgresql://postgres:postgres@localhost:5432/Homework9')
connection = engine.connect()
```

```python
#Import all tables associated with database
#parse_dates so they're in the correct date_time format
employees = pd.read_sql('select * from "Employees"', connection, parse_dates=['birth_date', 'hire_date'])
# employees
```

```python
#Check to see if null values are present, if there are then replace them
employees.isnull().values.any()
```

```python
#Salaries table
salary = pd.read_sql('select * from "Salaries"', connection)
# salary
```

```python
#Check to see if null values are present, if there are then replace them
salary.isnull().values.any()
```

```python
#Titles table
titles = pd.read_sql('select * from "Titles"', connection)
# titles
```

```python
#Check to see if null values are present, if there are then replace them
titles.isnull().values.any()
```

<font size="5">Create a histogram to visualize the most common salary ranges for employees.</font>

```python
#Combine data into a single dataset
salary_ranges = pd.merge(employees, salary, on="emp_no", how="outer")
# salary_ranges
```

```python
# salary_ranges.dtypes
```

```python
#Pandas method histogram chart
salary_ranges.hist('salary')
plt.savefig("Images/SalaryRangeHistPandas.png")
```

```python
#Matplotlib.pyplot method histogram chart
ax = salary_ranges.hist(column='salary', bins=25, grid=False, figsize=(12,8), color='indigo', zorder=2, rwidth=0.9)
ax = ax[0]

for x in ax:
    
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)
        
x.set_title("Most Common Salary Ranges for Employees", weight='bold', size=14)
x.set_xlabel("Salary Ranges", labelpad=20, weight='bold', size=12)
x.set_ylabel("Salary", labelpad=20, weight='bold', size=12)
plt.savefig("Images/SalaryRangeHistPyPlot.png")
```

<font size="5">Create a bar chart of average salary by title.</font>

```python
#Rename column title_id to emp_title_id for merge
rename_title_id = titles.rename(columns={"title_id": "emp_title_id"})
# rename_title_id
```

```python
#Combine data into a single dataset
salary_title = pd.merge(salary_ranges, rename_title_id, on="emp_title_id", how="outer")
# salary_title
```

```python
#Use the group by method on the values in 'title', 'salary', and find the average
group_title = salary_title.groupby(['title'])['salary'].mean()
# group_title
```

```python
#Create a bar chart using Pandas method
panda_chart = group_title.plot(kind='bar', color='teal')
panda_chart.set_xlabel("Title", labelpad=10, weight='bold', size=10)
panda_chart.set_ylabel("Average Salary", labelpad=10, weight='bold', size=10)
plt.title("Average Salary by Title", weight='bold', size=12)
plt.savefig("Images/AvgSalaryBarChartPandas.png")
plt.show()
```

```python
#Generate a bar plot using Pyplot
x_axis=np.arange(len(group_title))
plt.figure(figsize=(6,4))
plt.bar(x_axis, group_title, color='lightcoral', alpha=0.5, align="center")

# Tell matplotlib where we would like to place each of our x axis headers using list comprehension
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Assitance Engineer", "Engineer", "Manager", "Senior Engineer", "Senior Staff", "Staff", "Technique Leader"], rotation="vertical")

# Set a Title and labels
plt.title("Average Salary by Title", weight='bold', size=12)
plt.xlabel("Title", labelpad=10, weight='bold', size=10)
plt.ylabel("Average Salary", labelpad=10, weight='bold', size=10)
plt.savefig("Images/AvgSalaryBarChartPyplot.png")
plt.show()
```
